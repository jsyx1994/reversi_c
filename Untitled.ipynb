```python
import tensorflow as tf
import numpy as np
train_data = np.genfromtxt('/Users/mro/Desktop/cnnInput',
                               delimiter=' ', dtype=np.float32)
```

```python
train_data = train_data.reshape(-1, 64, 9)
```

```python
train_data.shape
```

```python
train_labels = np.genfromtxt('/Users/mro/Desktop/cnnOut',
                                 dtype=np.float32)
```

```python
train_labels.shape
```

```python
data = tf.data.Dataset.from_tensor_slices((train_data, train_labels))
data = data.repeat().batch(64)
x = data.make_one_shot_iterator().get_next()
```

```python
sess = tf.Session()
```

```python
sess.run(x)
```
